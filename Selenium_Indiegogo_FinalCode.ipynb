# Selenium Test

```python
from selenium import webdriver
import time
import csv
import pandas as pd

# Libraries for Explicit Waits
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
path = "C:\\Users\\Wewe\\chromedriver"
driver = webdriver.Chrome(path)
```

```python
driver.get("https://www.indiegogo.com/explore/all?project_type=campaign&project_timing=all&sort=trending")
time.sleep(5)
```

## Code for clicking on "Show More" button

```python
# Code to click on "Show More" button
for x in range(60):
    try:
        # Locate the Show More button
        show_more = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.LINK_TEXT, "SHOW MORE"))
        )
        show_more.click()
        time.sleep(3)

    except:
        print("Error")
```

## Code for collecting the links from all projects within the page

```python
# Code to extract the urls of all projects that is loaded
links = driver.find_elements(by=By.CSS_SELECTOR, value="div.discoverableCard > a")
print(links)
link_list = []
for link in links:
    href = link.get_attribute("href")
    link_list.append(href)
```

## Code for Data Extraction

```python
# Append the results to Dictionary
data = {
    'status': [],
    'title': [],
    'tagline': [],
    'creator': [],
    'num_campaign': [],
    'location': [],
    'amount_funded': [],
    'amount_backers': [],
    'percentage': [],
    'goal_type': [],
    'days_left': [],
    'story': [],
    'tags': [],
#    'faq': [],
#    'updates': [],
    'discussion': []
    }

# Code to loop and visit every project page
loop_number = 0
for url in link_list:
    driver.get(url)
    time.sleep(2)

    # Code to extract data from basic section
    try:
        # Basics Section data (status, title & tagline)
        basicssection = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, 'basicsSection'))
        )
        basicssection_text = basicssection.text
        basicssection_text = basicssection_text.split("\n")
        status = basicssection_text[0]
        title = basicssection_text[1]
        tagline = basicssection_text[2]
    except:
        status = "Error"
        title = "Error"
        tagline = "Error"

    try:
        # Creator
        creator = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "basicsCampaignOwner-details-name"))
        )
        creator = creator.text
        
        # Number of Campaigns
        num_campaign = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "basicsCampaignOwner-details-count"))
        )
        num_campaign = num_campaign.text

        # Location
        location = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "basicsCampaignOwner-details-city"))
        )
        location = location.text

        # Amount funded
        amount_funded = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "basicsGoalProgress-amountSold"))
        )
        amount_funded = amount_funded.text

        # Number of backers
        amount_backers = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "basicsGoalProgress-claimedOrBackers"))
        )
        amount_backers = amount_backers.text
        
    except:
        creator = "Error"
        num_campaign = "Error"
        location = "Error"
        amount_funded = "Error"
        amount_back = "Error"
    
    try:
        # Percentage funded & Total fund goal
        percentage = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "basicsGoalProgress-progressDetails-detailsGoal-goalPercentageOrInitiallyRaised"))
        )
        percentage = percentage.text

        # Goal type
        goal_type = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "basicsGoalProgress-progressDetails-detailsGoal-goalWording"))
        )
        goal_type = goal_type.text
    except:
        if status == "INDEMAND":
            percentage = "not_available"
            goal_type = "not_available"
        else:
            percentage = "Error"
            goal_type = "Error"

    try:
        # Days left
        days_left = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "basicsGoalProgress-progressDetails-detailsTimeLeft"))
        )
        days_left = days_left.text
    
    except:
        if status in ["INDEMAND", "CLOSED"]:
            days_left = "not_available"
        else:
            days_left = "Error"

    # Code to extract data from story and tags
    try:
        # Tags data
        tags = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "bottomAssignedTags"))
        )
        tags = tags.text

        # Story data
        story = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "routerContentStory"))
        )
        story = story.text

    except:
        story = 'Error'
        tags = 'Error'

    # # Code to extract data from faq # This code is not relevant for the research
    # try:
    #     # Navigate to faq tab and extracting faq data
    #     link_faq = WebDriverWait(driver, 10).until(
    #         EC.presence_of_element_located((By.LINK_TEXT, "FAQ"))
    #     )
    #     link_faq.click()
    #     time.sleep(2)
    #     faq = WebDriverWait(driver, 20).until(
    #         EC.presence_of_element_located((By.CLASS_NAME, "routerContentFaq"))
    #     )
    #     print(faq.text)
    #     faq = faq.text
    #     print("-----------------------------------------------------------------")

    # except:
    #     faq = "Error"

    # # Code to extract data from updates # This code is not relevant for the research
    # try:
    #     # Navigate to updates tab and extracting updates data
    #     link_updates= WebDriverWait(driver, 10).until(
    #         EC.presence_of_element_located((By.PARTIAL_LINK_TEXT, "UPDATES"))
    #     )
    #     link_updates.click()
    #     time.sleep(2)
    #     updates = WebDriverWait(driver, 20).until(
    #         EC.presence_of_element_located((By.CLASS_NAME, "routerContentUpdates"))
    #     )
    #     print(updates.text)
    #     updates = updates.text
    #     print("-----------------------------------------------------------------")

    # except:
    #     updates = "Error"

    # Code to extract data from discussion
    try:
        # Navigate to discussion tab and extracting discussion data
        link_discussion = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.PARTIAL_LINK_TEXT, "DISCUSSION"))
        )
        driver.execute_script("arguments[0].scrollIntoView(true);", link_discussion)
        link_discussion.click()
        time.sleep(2)
        discussion = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.CLASS_NAME, "routerContentSection-content"))
        )
        discussion = discussion.text

    except:
        discussion = "Error"
    
    loop_number += 1
    print("Collecting data variables from project", loop_number)
    
    # Append the results to CSV file
    data['status'].append(status)
    data['title'].append(title)
    data['tagline'].append(tagline)
    data['creator'].append(creator)
    data['num_campaign'].append(num_campaign)
    data['location'].append(location)
    data['amount_funded'].append(amount_funded)
    data['amount_backers'].append(amount_backers)
    data['percentage'].append(percentage)
    data['goal_type'].append(goal_type)
    data['days_left'].append(days_left)
    data['story'].append(story)
    data['tags'].append(tags)
    #data['faq'].append(faq)
    #data['updates'].append(updates)
    data['discussion'].append(discussion)
```

```python
# Code for extracting the numeric or decimal values from the string

# Percentage Clean
# Replacing the 'not_available' data points for 100%, since they only occur in INDEMAND projects which have guaranteed at least 100% funding
df['percentage'] = df['percentage'].replace('not_available', "100")
# Create a new column 'percentage_decimal' to store the converted values
df['percentage_decimal'] = df['percentage'].str.extract(r'(\d+)').astype(float) / 100

# Amount of Backers Clean
# Create a new column 'backers_numeric' to store the extracted numerical values
df['backers_numeric'] = df['amount_backers'].str.extract(r'(\d+)').astype(int)

# Number of Campaign Clean
# Create a new column 'backers_numeric' to store the extracted numerical values
df['campaigns_numeric'] = df['num_campaign'].str.extract(r'(\d+)').astype(int)

# Amount Funded Clean
# Remove commas from 'amount_funded' column
df['amount_funded'] = df['amount_funded'].str.replace(',', '')
# Create a new column 'price_numeric' to store the values without the euro sign
df['funded_numeric'] = df['amount_funded'].str.replace('€', '').astype(int)

# Days Left Clean
# Create a new column 'days_numeric' to store the extracted numeric values
df['days_numeric'] = df['days_left'].str.extract(r'(\d+)').fillna(0)
# Search for columns with 'hours' and assign the numeric value 1
df.loc[df['days_left'].str.contains('hours'), 'days_numeric'] = 1
df['days_numeric'] = df['days_numeric'].astype(int)

# Status Clean
# Create a new column 'success' to store the binary values
df['success'] = 0
# Assigning 1 to the projects that reached their funding goal. The projects that failed already have a 0 assigned.
df.loc[df['status'] == 'INDEMAND', 'success'] = 1
df.loc[(df['status'] == 'FUNDING') & (df['percentage_decimal'] >= 1), 'success'] = 1
```

```python
# Splitting Tags into Main Tags and Secondary Tags

# Create separate lists for main tags and secondary tags
main_tags = ['TECH & INNOVATION', 'CREATIVE WORKS', 'COMMUNITY PROJECTS']
secondary_tags = [
    'AUDIO', 'CAMERA GEAR', 'EDUCATION', 'ENERGY & GREEN TECH', 'FASHION & WEARABLES',
    'FOOD & BEVERAGES', 'HEALTH & FITNESS', 'HOME', 'PHONES & ACCESSORIES', 'PRODUCTIVITY',
    'TRANSPORTATION', 'TRAVEL & OUTDOORS', 'ART', 'COMICS', 'DANCE & THEATER', 'FILM', 'MUSIC',
    'PHOTOGRAPHY', 'PODCASTS, BLOGS & VLOGS', 'TABLETOP GAMES', 'VIDEO GAMES',
    'WEB SERIES & TV SHOWS', 'WRITING & PUBLISHING', 'CULTURE', 'ENVIRONMENT', 'HUMAN RIGHTS',
    'LOCAL BUSINESSES', 'WELLNESS'
]

# Create new columns 'main_tag' and 'secondary_tag' and initialize them with None
df['main_tag'] = 'no_main_tag'
df['secondary_tag'] = 'no_secondary_tag'

# Iterate over the DataFrame rows
for index, row in df.iterrows():
    tags = row['tags']
    
    # Check for main tags and update 'main_tag' column
    for tag in main_tags:
        if tag in tags:
            df.at[index, 'main_tag'] = tag
    
    # Check for secondary tags and update 'secondary_tag' column
    for tag in secondary_tags:
        if tag in tags:
            df.at[index, 'secondary_tag'] = tag
```

```python
# Creating a list of keywords for References on Prior Work
works = [
    'adapted from',
    'based on',
    'borrows elements from',
    'borrowed from',
    'built upon',
    'citated by',
    'citated from',
    'derived from',
    'draws from',
    'draws upon',
    'echoes',
    'expands upon',
    'in homage to',
    'in the style of',
    'influenced by',
    'influenced by the style of',
    'inspiration from',
    'inspired by',
    'inspired by the works of',
    'incorporates elements of',
    'modeled after',
    'pays homage to',
    'references by',
    'references from',
    'resembles the works of',
    'reminiscent of',
    'takes cues from'
]

# Creating a code for matching and counting the references

# Initialize 'num_references' column with zeros
df['num_references'] = 0
df['desc_references'] = ''

# Iterate over the works list and increment 'num_references' and build 'desc_references' list for each occurrence
for work in works:
    pattern = rf'([^\.]*{re.escape(work)}[^\.]*)'
    match = df['story'].str.contains(work, flags=re.IGNORECASE)
    df.loc[match, 'num_references'] += 1
    df.loc[match, 'desc_references'] += df.loc[match, 'story'].str.extract(pattern, expand=False) + ', '

# Remove trailing comma and whitespace in 'desc_references'
df['desc_references'] = df['desc_references'].str.rstrip(', ')

# Fill the NaNs in 'desc_references'
df['desc_references'] = df['desc_references'].fillna('')

# Cleaning the column desc_references
# Replacing \n into white space
df['desc_references'] = df['desc_references'].str.replace(".\n", " ", regex=True)
df['desc_references'] = df['desc_references'].str.replace("\n", " ", regex=True)

# Creating a column for binary values for references, 0 for no references, 1 for containing 1 or more references
df['references_binary'] = (df['num_references'] > 0).astype(int)
```

```python
# Cleaning the Discussion text
patterns = ["You must be a backer to join the discussion. Log in or sign up.",
            "No Comments",
            "No discussion yet",
            "Log in or sign up to join the discussion"]

for pattern in patterns:
    df['discussion'] = df['discussion'].str.replace(pattern, "", regex=True)

# Replacing \n into white space
df['discussion'] = df['discussion'].str.replace(".\n", " ", regex=True)
df['discussion'] = df['discussion'].str.replace("\n", " ", regex=True)
```

```python
# Code to create and output CSV file
df.to_csv('scrape_output_cleaned', sep=';')
```
